```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re 

url = 'https://optioncharts.io/options/QQQ'
driver = webdriver.Chrome()
driver.get(url)
html_content = driver.page_source
soup = BeautifulSoup(html_content, 'html.parser')
op_stats = soup.find_all('div', class_='tw-font-semibold')
op_stat_names = soup.find_all('div', class_='tw-text-sm tw-text-gray-500')
stat_strings = [x.get_text().rstrip() for x in op_stat_names]
raw_stats = [x.get_text() for x in op_stats]
driver.quit()
stats = []
metrics = {}
for stat in raw_stats:
    stats.append(re.sub(r'[^\d,.%]', '', stat))
for k, v in zip(stat_strings, stats):
    metrics[k] = v
metrics['IV High'] = metrics['IV High'].split('%')[0]
metrics['IV Low'] = metrics['IV Low'].split('%')[0]
# print(len(stat_strings), len(iwm_stats))
print(metrics)
```
